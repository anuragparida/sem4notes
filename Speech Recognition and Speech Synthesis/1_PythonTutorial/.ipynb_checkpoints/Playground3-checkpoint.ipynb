```python
import numpy as np

# classes
class CCircle(object):

    def __init__(self, InitialRadius = 0.0):
        self.__radius = InitialRadius

    def GetRadius(self):
        return self.__radius

    def SetRadius(self, NewRadius):
        assert NewRadius >= 0.0, 'NewRadius must be greater or equal zero'
        self.__radius = NewRadius

    def EvaluateArea(self):
        return np.pi*(self.__radius**2)

ACircle = CCircle()
ACircle.SetRadius(2.0)
print('area = ', ACircle.EvaluateArea())

AListOfCircles = []
for i in range(10):
    AListOfCircles.append(CCircle(i))

for i in range(len(AListOfCircles)):
    print('area = ', AListOfCircles[i].EvaluateArea())

for ACircle in AListOfCircles:
    print('area = ', ACircle.EvaluateArea())
```

```python
class CCar(object):

    def __init__(self, FuelConsumptionPer100km = 6.3):
        self.__FuelConsumptionPer100km = FuelConsumptionPer100km
        self.__FueltankSize = 100
        self.__DrivenDistanceInkm = 0.0
        self.__FuelLevel = 0.0

    def DriveDistance(self, TargetDistanceInkm):
        TargetFuelConsumption = TargetDistanceInkm / 100 * self.__FuelConsumptionPer100km
        if TargetFuelConsumption > self.__FuelLevel:
            TargetFuelConsumption = self.__FuelLevel
            TargetDistanceInkm = self.__FuelLevel / self.__FuelConsumptionPer100km * 100
        self.__FuelLevel -= TargetFuelConsumption
        self.__DrivenDistanceInkm += TargetDistanceInkm
        
    def AddFuel(self, AdditionalFuel):
        self.__FuelLevel += AdditionalFuel
        if self.__FuelLevel > self.__FueltankSize:
            self.__FuelLevel = self.__FueltankSize

    def GetAmountOfFuel(self):
        return self.__FuelLevel

    def GetDrivenDistanceInkm(self):
        return self.__DrivenDistanceInkm

    #def SetFuelConsumptionPer100km(self, FuelConsumptionPer100km):
    #    self.__FuelConsumptionPer100km = FuelConsumptionPer100km

    def MakeNoise(self):
        return 'hup'

    def GetColour(self):
        return 'blue'

    def GetNumberOfSeats(self):
        return 7

ACar = CCar()
ACar.AddFuel(500)
ACar.DriveDistance(230000)
print('current fuel = ', ACar.GetAmountOfFuel())
print('driven distance = ', ACar.GetDrivenDistanceInkm())

'''
the class CCar has now 2 tasks:
a) fuel, driving, adding fuel
b) look and feel of the car
'''
```

```python
class CDrivingAndFuelFunctionality(object):
    
    def __init__(self, FuelConsumptionPer100km = 6.3):
        self.__FuelConsumptionPer100km = FuelConsumptionPer100km
        self.__FueltankSize = 100
        self.__DrivenDistanceInkm = 0.0
        self.__FuelLevel = 0.0

    def DriveDistance(self, TargetDistanceInkm):
        TargetFuelConsumption = TargetDistanceInkm / 100 * self.__FuelConsumptionPer100km
        if TargetFuelConsumption > self.__FuelLevel:
            TargetFuelConsumption = self.__FuelLevel
            TargetDistanceInkm = self.__FuelLevel / self.__FuelConsumptionPer100km * 100
        self.__FuelLevel -= TargetFuelConsumption
        self.__DrivenDistanceInkm += TargetDistanceInkm
        
    def AddFuel(self, AdditionalFuel):
        self.__FuelLevel += AdditionalFuel
        if self.__FuelLevel > self.__FueltankSize:
            self.__FuelLevel = self.__FueltankSize

    def GetAmountOfFuel(self):
        return self.__FuelLevel

    def GetDrivenDistanceInkm(self):
        return self.__DrivenDistanceInkm

# cohesion
# a high value for cohesian corresponds to
# every class variable is used in nearly every procedure

class CLookNFeel(object):

    def MakeNoise(self):
        return 'hup'

    def GetColour(self):
        return 'blue'

    def GetNumberOfSeats(self):
        return 7

class CCar(object):

    def __init__(self):
        self.__DrivingAndFuelFunctionality = CDrivingAndFuelFunctionality()
        self.__LookNFeel = CLookNFeel()
```

```python
import numpy as np

class CGeometricalObject(object):

    def __init__(self):
        self._area = 0.0
        self._surrounding = 0.0

    def GetArea(self):
        return self._area

    def GetSurrounding(self):
        return self._surrounding

class CCircle(CGeometricalObject):

    def __init__(self, Radius = 1.0):        
        super().__init__()
        self.SetRadius(Radius)

    def SetRadius(self, NewRadius):
        self.__Radius = NewRadius
        self._area = NewRadius**2 * np.pi
        self._surrounding = NewRadius * np.pi * 2

class CSquare(CGeometricalObject):

    def __init__(self, EdgeLength = 1.0):
        self.__EdgeLength = EdgeLength
        super().__init__()
        self._area = EdgeLength**2
        self._surrounding = 4*EdgeLength
    

ACircle = CCircle(3.0)
print(ACircle.GetArea())
print(ACircle.GetSurrounding())
ACircle.SetRadius(5)
print(ACircle.GetArea())
print(ACircle.GetSurrounding())

ASquare = CSquare(5)
print(ASquare.GetArea())
print(ASquare.GetSurrounding())
```
