# Channel model
The basic concept of a denoising algorithm is a channel model with input signal $x(n)$ and output signal $y(n)$. The channel models all kind of degradation, e.g. additive noise, echos, non-linear distortions, and so on.

In the following, we will assume a simple channel model with additive noise $d(n)$ and linear filtering $h(n)$:

$y(n)=h(n)*x(n)+d(n)$

This channel introduces two effects:

1) linear filtering by convolution with the impulse response $h(n)$ and
2) additive noise by the operation $+d(n)$.

Denoising is basically the inversion of all unwanted effects of the channel.

## Audio data
The Oldenburger sentence test is often used for testing speech intelligibility or hearing losses.
Each sentence consists of five words. For each of the five positions, there exists ten possible words.
This leads to a set of $5^{10}\approx 9.77\cdot 10^6$ possible combinations.

In the following code block a sample sentence is constructed randomly.

```python
import whisper
import torch
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os
os.chdir('../Python')

SamplingRateWhisper = 16000 # the sampling rate should always be 16 kHz

WordList0 = ['Peter', 'Kerstin', 'Tanja', 'Ulrich', 'Britta', 'Wolfgang', 'Stefan', 'Thomas', 'Doris', 'Nina']
WordList1 = ['bekommt', 'sieht', 'kauft', 'gibt', 'schenkt', 'verleiht', 'hat', 'gewann', 'nahm', 'malt']
WordList2 = ['drei', 'neun', 'sieben', 'acht', 'vier', 'fünf', 'zwei', 'achtzehn', 'zwölf', 'elf']
WordList3 = ['große', 'kleine', 'alte', 'nasse', 'schwere', 'grüne', 'teure', 'schöne', 'rote', 'weiße']
WordList4 = ['Blumen', 'Tassen', 'Autos', 'Bilder', 'Dosen', 'Sessel', 'Messer', 'Schuhe', 'Steine', 'Ringe']
ListOfLists = [WordList0, WordList1, WordList2, WordList3, WordList4]

def GetSentenceString():
    s = ''
    for n in range(len(ListOfLists)):
        idx = np.random.randint(len(ListOfLists[n]))
        s += ListOfLists[n][idx]
        if n < (len(ListOfLists) - 1):
            s += ' '
        else:
            s += '.'
    return s

print(GetSentenceString())
```

## Generating the audio data
In the following code block a Text To Speech (TTS) System is used to create the audio file.

```python
import pyttsx3

Filename = 'OldenburgerSatztest.wav'
SentenceAsString = GetSentenceString()

# initialize the speech synthesizer
engine = pyttsx3.init()

rate = engine.getProperty('rate')   # getting details of current speaking rate
engine.setProperty('rate', 125)     # setting up new voice rate

# select the language
engine.setProperty('voice', 'german')

engine.save_to_file(SentenceAsString, Filename)

# wait until everything is finished
engine.runAndWait()
```

## White noise
The simplest noise model is additive white gaussian noise. Gaussian noise has a normal distribution in time domain. White noise has nearly equal power density spectrum over all frequencies.

```python
def GenerateWhiteGaussianNoise(NumberOfSamples):
    # mean value = 0
    # standard deviation 1
    return np.random.randn(NumberOfSamples)
```

## Apply the channel model
The channel model is applied in two steps:

1) $y(n)=h(n)*x(n)$
2) $z(n)=y(n)+d(n)$

```python
import wave
with wave.open(Filename, 'rb') as wave_datei:
    rohdaten = wave_datei.readframes(wave_datei.getnframes())
    Fs = wave_datei.getframerate()
    x = np.frombuffer(rohdaten, dtype=np.int16) / (2**15)

plt.plot(x)
plt.show()

h = np.zeros((10))
h[0] = 1
y = np.convolve(x, h)

d = GenerateWhiteGaussianNoise(y.shape[0])

TargetSNR = 30
noise_gain = np.sqrt(np.sum(y**2)/(np.sum(d**2)*(10**(TargetSNR/10))))
z = y + noise_gain*d
SNR = 10*np.log10(np.sum(y**2)/np.sum((z-y)**2))
assert np.abs(SNR-TargetSNR) < 1e-2, 'wrong SNR'
```

## Levenshtein distance
A first simple measurement of the speech intelligibility is the Levenshtein distance between the original sentence and the output of a Speech To Text (STT) System, e.g. Whisper.

The Levenshein distance between "3" and "drei" is: distance$=4$. For a human listener, the distance would be zero, which corresponds to a correct output.
Therefore, a set of rules must be applied in order to replace ambiguous outputs of the STT.

```python
from Levenshtein import distance as lev

ModelSize = "small" # ["tiny", "base", "small", "medium", "large"]
Language = 'German' # ["German", "English"]
devices = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") # source: https://github.com/openai/whisper/discussions/373
model = whisper.load_model(ModelSize, device = devices)  

def DoResampling(x, OriginalSamplingRate):
    if np.abs(SamplingRateWhisper - OriginalSamplingRate) > 1:
        y = librosa.resample(x, orig_sr = OriginalSamplingRate, target_sr = SamplingRateWhisper)
        y *= (np.max(np.abs(x)) / np.max(np.abs(y)))
        return y
    else:
        return x

def CallWhisper(x, OriginalSamplingRate):
    y = DoResampling(x, OriginalSamplingRate)
    result = model.transcribe(y.astype(np.float32), language=Language, verbose=0)
    return result["text"]

def ApplySetOfRules(InputString):
    InputString = InputString.strip()
    InputString = InputString.replace("3","drei")
    InputString = InputString.replace("9","neun")
    InputString = InputString.replace("7","sieben")
    InputString = InputString.replace("8","acht")
    InputString = InputString.replace("4","vier")
    InputString = InputString.replace("5","fünf")
    InputString = InputString.replace("2","zwei")
    InputString = InputString.replace("18","achtzehn")
    InputString = InputString.replace("12","zwölf")
    InputString = InputString.replace("11","elf")
    return InputString

TextDetected = CallWhisper(z, Fs)
TextDetected = ApplySetOfRules(TextDetected)

print('input: ', SentenceAsString)
print('output: ', TextDetected)
print('Levenshtein distance: ', lev(SentenceAsString, TextDetected))
```

## Speech Transmission Index
The speech transmission index (STI) is an additional measurement for the speech intelligibility. The STI is in the range from "0", which corresponds to no intelligibility, to "1" which corresponds to perfect intelligibility.

Here, an implementation of 'A method to determine the speech transmission index from speech waveforms' is used. 

```python
import sti

stis = sti.stiFromAudio(x, z, Fs)
print(stis)
```

## Correlation between STI and Levenshtein distance
In the following code, the correlation between the STI and the Levenshtein distance for different noise levels is evaluated.

The following can be derived:
1) The worse the SNR, the higher the Levenshtein distance.
2) The worse the SNR, the lower the STI.

```python
LevenshteinList = []
STIList = []
for TargetSNR in [-15, -10, -5, 0, 5, 10, 15]:
    d = GenerateWhiteGaussianNoise(y.shape[0])
    noise_gain = np.sqrt(np.sum(y**2)/(np.sum(d**2)*(10**(TargetSNR/10))))
    z = y + noise_gain*d
    TextDetected = CallWhisper(z, Fs)
    TextDetected = ApplySetOfRules(TextDetected)
    LevenshteinList.append(lev(SentenceAsString, TextDetected))
    STIList.append(sti.stiFromAudio(x, z, Fs))

plt.scatter(STIList, LevenshteinList)
plt.show()

def PearsonCorrelation(x, y):
    x_0m = x - np.mean(x)
    y_0m = y - np.mean(y)
    return np.sum(x_0m * y_0m) / np.sqrt(np.sum(x_0m**2) * np.sum(y_0m**2))

def PearsonRankCorrelation(x, y):
    return PearsonCorrelation(np.argsort(x), np.argsort(y))

print('Pearson correlation = ', PearsonCorrelation(np.array(STIList), np.array(LevenshteinList)))
print('Pearson rank correlation = ', PearsonRankCorrelation(np.array(STIList), np.array(LevenshteinList)))
```

## Programming exercise
For evaluating speech intelligibility, usually a noise type is used, which has an average power density spectrum equal to human speech. Because, this noise has not equal power density spectrum over all frequencies, it is no longer called white noise. Instead it is so called coloured noise.

In the paper 'An international comparison of long-term average speech spectra' the average spectrum of human noise is published as a set of frequencies and their corresponding Level.

The procedure GenerateNoiseWithLongTermAverageSpeechSpectrum generates noise with an average spectrum comparable to human voice. For this, the following three procedures should be implemented:

1) Program the procedure GenerateBandpassNoise by defining a normal distributed random variable and filter it with a bandpass with given cutoff frequencies lowcut and highcut.

2) Program the procedure CenterFrequencyToLowcutHighcut by transforming the center frequency to the corresponding cutoff frequencies lowcut and highcut. The difference between lowcut and highcut shoud be $f_\Delta$. The center frequency should be the mean value of lowcut and highcut.

3) Program the procedure LevelToLinearMagnitude by inverting the function $L=20\cdot\log_{10}a$.
