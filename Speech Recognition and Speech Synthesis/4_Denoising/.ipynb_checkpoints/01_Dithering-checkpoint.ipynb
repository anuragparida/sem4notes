# Dithering

Dithering adds artificial noise to a recorded signal. Dithering circumvents certain numerical problems in the context of quantisation.

Assuming a recording with a large range of levels from highest amplitude (in this example 2) to lowest amplitude (in this example 0).

```python
import numpy as np
import matplotlib.pyplot as plt

# define the signal
Fs = 48000
f = 1
n = np.arange(Fs)
x = np.cos(2*np.pi*f*n/Fs) + 1

# quantisation
w = 4
Delta = (np.max(x) - np.min(x))/(2**w)
x_tilde = Delta * np.round(x/Delta)

plt.plot(x, 'b')
plt.plot(x_tilde, 'r')
plt.show()
```

A typical task in audio engineering is to evaluate the level as a logarithmic value of the root mean square (RMS):

```python
def EvaluateLevel(x):
    BlocksizeInSamples = 100
    Level_x       = np.zeros(int(x.shape[0]/BlocksizeInSamples))
    for k in range(Level_x.shape[0]):
        idx1 = k*BlocksizeInSamples
        idx2 = idx1 + BlocksizeInSamples
        Level_x[k] = 20*np.log10(np.sqrt(np.mean(x[idx1:idx2]**2)) + 1e-10)
    return Level_x

Level_x       = EvaluateLevel(x)
Level_x_tilde = EvaluateLevel(x_tilde)
    
plt.plot(Level_x, 'b')
plt.plot(Level_x_tilde, 'r')
plt.show()

print('mean level of x: ', np.mean(Level_x), ' dB')
print('mean level of x_tilde: ', np.mean(Level_x_tilde), 'dB')
```

Obviously, in the signal $\tilde{x}(n)$ some values are quantised to digital zero.
These quantisation values of exactly zero leads
a) to numerical problems, because the logarithm of digital zero is minus infinity and
b) to misleading interpretation if typical audio features like the mean level.

A typical description of this problem is:
The quantisation to digital zero takes away the quantisation noise.
In order to improve numerical stability, one workaround is dithering. When the audio signal is loaded, the dithering noise $d(n)$ is added:

$-\frac{\Delta}{2}<d(n)<\frac{\Delta}{2}$

```python
d = (np.random.rand(x_tilde.shape[0]) - 0.5) * Delta
x_dithered = x_tilde + d

Level_x_dithered = EvaluateLevel(x_dithered)

assert np.sum(np.abs(x_tilde - Delta*np.round(x_dithered/Delta))) < 1e-10, 'dithering changes the quantised values'

plt.plot(Level_x, 'b')
plt.plot(Level_x_tilde, 'r')
plt.plot(Level_x_dithered, 'k')
plt.show()

print('mean level of x: ', np.mean(Level_x), ' dB')
print('mean level of x_tilde: ', np.mean(Level_x_tilde), 'dB')
print('mean level of x_dithered: ', np.mean(Level_x_dithered), 'dB')
```

## Exercise:

Adding two independent uniformly distributed variables leads to a variable with a triangle probability density function.

Generate a random signal $d_\Delta$ with triangle probability distribution in the range $-\frac{\Delta}{2}<d(n)<\frac{\Delta}{2}$.

Plot the histogram of this signal $d_\Delta$.

```python
### solution
x_uniform1 = np.random.rand(10000) # 0...1, uniform distribution
x_uniform2 = np.random.rand(x_uniform1.shape[0]) # 0...1, uniform distribution
x_triangle = x_uniform1 + x_uniform2 # 0...2, triangle distribution
x_triangle *= 0.5 # 0...1, triangle distribution
x_triangle -= 0.5 # -0.5...0.5, triangle distribution

plt.hist(x_triangle)
plt.show()
```
