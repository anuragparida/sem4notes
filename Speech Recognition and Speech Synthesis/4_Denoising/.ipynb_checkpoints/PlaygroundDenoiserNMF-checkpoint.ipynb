```python
import matplotlib.pyplot as plt
import numpy as np
import webrtcvad
import os
os.chdir('../Python')
import WaveInterface
import Constants
import TrainingsDataInterface

eps = 1e-15
TargetSamplingRate = Constants.theConstants.getSamplingFrequencyMicrofone()
```

```python
NoiseFilenames = []
NoiseFilenames.append('Noise/FeuerwehrautoMessmikrofon.wav')
NoiseFilenames.append('Noise/Getr√§nkeautomatMessmikrofon.wav')
NoiseFilenames.append('Noise/HydraulikzylinderMessmikrofon.wav')
NoiseFilenames.append('Noise/KompressorMessmikrofon.wav')

def GetNoiseFile():
    Index = np.random.randint(len(NoiseFilenames))
    x, Fs, bits = WaveInterface.ReadWave(NoiseFilenames[Index])
    assert Fs == TargetSamplingRate, 'Samplingrate mismatch'
    return x

ATrainingsDataInterface = TrainingsDataInterface.CTrainingsDataInterface()
def GetSpeech(NumberOfSamplesNoise):
    x = np.random.randn(NumberOfSamplesNoise) * eps
    StartIndex = 0
    while StartIndex < NumberOfSamplesNoise:
        CommandIndex = np.random.randint(ATrainingsDataInterface.GetNumberOfCommands())
        InstanceIndex = np.random.randint(ATrainingsDataInterface.GetNumberOfCommandInstances(CommandIndex))
        y, Fs, bits = ATrainingsDataInterface.GetWaveOfCommandInstance(CommandIndex, InstanceIndex)
        assert Fs == TargetSamplingRate, 'Samplingrate mismatch'
        StopIndex = StartIndex + y.shape[0]
        if StopIndex > NumberOfSamplesNoise:
            StopIndex = NumberOfSamplesNoise
            y = y[:NumberOfSamplesNoise - StartIndex]
        x[StartIndex:StopIndex] += y
        StartIndex = StopIndex
    return x

d = GetNoiseFile()
x = GetSpeech(d.shape[0])
assert x.shape[0] == d.shape[0], 'size mismatch'
```

```python
TargetSNR = 5

def ApplyNoise(x, d, TargetSNR):
    noise_gain = np.sqrt(np.sum(x**2)/(np.sum(d**2)*(10**(TargetSNR/10))))
    d *= noise_gain
    return x + d
    
y = ApplyNoise(x, d, TargetSNR)
RealSNR = 10*np.log10(np.sum(x**2) / np.sum((x-y)**2))
assert np.abs(RealSNR - TargetSNR) < 0.1, 'wrong noise level'
```

```python
def BetaDivergence(x, y, beta):
    if beta == 0:
        return np.sum(x/y - np.log(x/y) - 1)
    elif beta == 1:
        return np.sum(x*(np.log(x) - np.log(y)) + y - x)
    else:
        return np.sum((x**beta + (beta - 1) * (y**beta) - beta * x * (y**(beta-1))) / (beta * (beta - 1)))

class COnlineNMF(object):

    def __init__(self, beta, T, I):
        self.__X = None
        self.__beta = beta
        self.__T = T
        self.__I = I
    
    def __Initialize(self, K):
        self.__X = np.zeros((K, 0))
        self.__B = np.random.rand(K, self.__I) + eps
        self.__G = np.zeros((self.__I, 0))

    def __UpdateTimeAxis(self, X):
        self.__X = np.concatenate((self.__X, np.reshape(X, (X.shape[0], 1))), axis = 1)
        self.__G = np.concatenate((self.__G, np.random.rand(self.__G.shape[0], 1)), axis = 1)
        if self.__X.shape[1] > self.__T:
            self.__X = self.__X[:, :-1]
            self.__G = self.__G[:, :-1]
            assert self.__X.shape[1] == self.__T, 'wrong reduction of time axis'
        assert self.__X.shape[1] == self.__G.shape[1], 'time axis must be identical'
        assert self.__G.shape[0] == self.__I, 'wrong 0-th dimension of G'

    def __EvalXest(self):
        Xest = np.matmul(self.__B, self.__G)
        assert Xest.shape == self.__X.shape, 'wrong shape of Xest'
        return Xest

    def __EvalZeta(self):
        Xest = self.__EvalXest()
        Zeta1 = self.__X * (Xest**(self.__beta - 2))
        Zeta2 = Xest**(self.__beta-1)
        return Zeta1, Zeta2
    
    def __UpdateB(self):
        Zeta1, Zeta2 = self.__EvalZeta()
        self.__B *= (np.matmul(Zeta1, np.transpose(self.__G)) + eps) / (np.matmul(Zeta2, np.transpose(self.__G)) + eps)

    def __UpdateG(self):
        Zeta1, Zeta2 = self.__EvalZeta()
        self.__G *= (np.matmul(np.transpose(self.__B), Zeta1) + eps) / (np.matmul(np.transpose(self.__B), Zeta2) + eps)

    def __NormalizeBG(self):
        Xest0 = self.__EvalXest()
        for i in range(self.__B.shape[1]):
            EB = np.sum(self.__B[:, i]**2)
            EG = np.sum(self.__G[i, :]**2)
            a = np.sqrt(EB*EG)
            self.__B[:, i] *= a/EB
            self.__G[i, :] *= a/EG
        Xest1 = self.__EvalXest()
        assert 10*np.log10(np.sum(Xest0**2) / np.sum((Xest0 - Xest1)**2)) > 100, 'error in normalization'
    
    def ProcessNewColumn(self, X, NumberOfIterations = 1):
        if self.__X is None:
            self.__Initialize(K = X.shape[0])
        if not (X is None): self.__UpdateTimeAxis(X)
        #BetaDivergence1 = BetaDivergence(self.__X, self.__EvalXest(), self.__beta)
        for n in range(NumberOfIterations):
            self.__UpdateG()
            self.__UpdateB()        
            self.__NormalizeBG()
        #BetaDivergence2 = BetaDivergence(self.__X, self.__EvalXest(), self.__beta)
        #assert BetaDivergence2 < BetaDivergence1, 'convergence error'
        return self.__B, self.__G
    
I = 5
B = np.random.randn(10, 5)
G = np.random.randn(B.shape[1], 20)
X = np.matmul(B, G)
AOnlineNMF = COnlineNMF(beta = 2, T = X.shape[1], I = I)
for col in range(X.shape[1]):
    B, G = AOnlineNMF.ProcessNewColumn(X[:, col])
```

```python
d = GetNoiseFile()
x = GetSpeech(d.shape[0])
y = ApplyNoise(x, d, TargetSNR = 5)

def EvalBG(x):
    NMFMemory = np.inf
    I = 5
    beta = 2
    
    AOnlineNMF = COnlineNMF(beta = beta, T = NMFMemory, I = I)
    HopSizeInMs = 10
    hs = int(HopSizeInMs * TargetSamplingRate / 1000)
    ws = 2*hs
    w = (1-np.cos(2*np.pi*(np.arange(ws)+0.5)/ws)) / 2
    
    NumberOfCols = (d.shape[0] - ws) // hs + 1
    for n in range(NumberOfCols):
        idx1 = n*hs
        idx2 = idx1 + ws
        X = np.abs(np.fft.rfft(x[idx1:idx2] * w))
        B, G = AOnlineNMF.ProcessNewColumn(X, 0)
    B, G = AOnlineNMF.ProcessNewColumn(None, 100)
    return B, G

Bx, Gx = EvalBG(x)
Bd, Gd = EvalBG(d)

plt.plot(Bx)
plt.show()
plt.plot(np.abs(np.fft.rfft(np.transpose(Gx))))
plt.show()
plt.plot(Bd)
plt.show()
plt.plot(np.abs(np.fft.rfft(np.transpose(Gd))))
plt.show()
```
