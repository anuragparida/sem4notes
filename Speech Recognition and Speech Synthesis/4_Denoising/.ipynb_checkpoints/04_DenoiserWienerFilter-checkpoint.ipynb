# Wiener Filter:

The Wiener filter is a common algorithm to suppress the noise and a good example for working with the spectrogram.

In the following a signal with increasing frequency and an envelope corresponding to the sinc-function is assumed:

```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 48000
t = np.arange(Fs)/Fs
x = np.sinc(10*(t - np.median(t))) * np.cos(2*np.pi*50*(t**1.5))

plt.plot(t, x)
plt.show()
```

## Channel model / distortion model
In the following we assume additive white Gaussian noise (AWGN) without any linear filtering in the recording process:

$y(n)=x(n)+d(n)$

```python
d = np.random.randn(x.shape[0]) * 0.1
y = x + d

plt.plot(t, y)
plt.show()
```

The Wiener-Filter in its simplest implementation works in the spectral domain:

$H(f)=\frac{\left|X(f)\right|^2}{\left|X(f)\right|^2+\left|D(f)\right|^2}$

$Z(f) = Y(f)\cdot H(f)$

$\left|X(f)\right|^2$ is the so called power spectral density of $x(n)$.

$\left|D(f)\right|^2$ is the so called power spectral density of the noise process.

```python
hs = 2**9 # hopsize
ws = 2*hs # windowsize
w = np.sqrt((1-np.cos(2*np.pi*(np.arange(ws) + 0.5)/ws))*0.5)
z_withoutFilter = np.zeros(y.shape)
z_withFilter = np.zeros(y.shape)

idx1 = 0
idx2 = ws
while idx2 < y.shape[0]:
    # evaluate the spectra
    X = np.fft.fft(x[idx1:idx2] * w)
    D = np.fft.fft(d[idx1:idx2] * w)
    Y = np.fft.fft(y[idx1:idx2] * w)
    # evaluate the filter
    H = np.abs(X)**2/(np.abs(X)**2 + np.abs(D)**2)
    # apply the filter
    Z = Y*H
    #overlap add
    z_withFilter[idx1:idx2] += np.fft.ifft(Z).real * w
    z_withoutFilter[idx1:idx2] += np.fft.ifft(Y).real * w
    
    idx1 += hs
    idx2 += hs
    
SNR1 = 10*np.log10(np.sum(x**2)/np.sum((z_withoutFilter-x)**2))
SNR2 = 10*np.log10(np.sum(x**2)/np.sum((z_withFilter-x)**2))

print('SNR before Wiener filter = ', SNR1, ' dB')
print('SNR after Wiener filter = ', SNR2, ' dB')
```

Unfortunately, in a real measurement environment, the true signal $x$ and the true noise $n$ are unknown. Instead, only the mixture is known:

$y(n)=x(n)+d(n)$

In order to assume the noise level, stationarity of the noise is assumed.
This means, that the power spectra of the noise $\left|D(f)\right|^2$ is nearly constant over time. This assumpetion is valid, when the noise is not changing a lot over time.
If stationarity of the noise is assumed, the power spectra can be estimated by evaluating the mean value over the time axis of the spectrogram.

Additionally, it can be assumed, that the signal $x(n)$ and the noise $d(n)$ are uncorrelated. In this case, the following can be shown:

$\left|Y(f)\right|^2 \approx \left|X(f)\right|^2 + \left|D(f)\right|^2$

From this, the filter can be evaluated by:

$H(f)=\frac{\left|X(f)\right|^2}{\left|X(f)\right|^2 + \left|D(f)\right|^2}=\frac{\left|Y(f)\right|^2 - \left|D(f)\right|^2}{\left|Y(f)\right|^2}=1-\frac{\left|D(f)\right|^2}{\left|Y(f)\right|^2}$

```python
hs = 2**9 # hopsize
ws = 2*hs # windowsize
w = np.sqrt((1-np.cos(2*np.pi*(np.arange(ws) + 0.5)/ws))*0.5)
z_withoutFilter = np.zeros(y.shape)
z_withFilter = np.zeros(y.shape)

a_max = 0.99
eps = 1e-10
counter = 0
D2 = 0.0
idx1 = 0
idx2 = ws
while idx2 < y.shape[0]:
    # evaluate the spectrum
    Y = np.fft.fft(y[idx1:idx2] * w)
    # evaluate the power spectrum
    Y2 = np.abs(Y)**2
    # estimate the power spectra of the noise by an IIR filter
    a = counter / (counter + 1)
    if a > a_max: a = a_max
    D2 = a * D2 + (1-a) * Y2
    counter += 1
    # evaluate the filter
    numerator = np.maximum(eps, Y2 - D2)
    denominator = Y2 + eps
    H = numerator / denominator
    # apply the filter
    Z = Y*H
    #overlap add
    z_withFilter[idx1:idx2] += np.fft.ifft(Z).real * w
    z_withoutFilter[idx1:idx2] += np.fft.ifft(Y).real * w
    
    idx1 += hs
    idx2 += hs
    
SNR1 = 10*np.log10(np.sum(x**2)/np.sum((z_withoutFilter-x)**2))
SNR2 = 10*np.log10(np.sum(x**2)/np.sum((z_withFilter-x)**2))

print('SNR before Wiener filter = ', SNR1, ' dB')
print('SNR after Wiener filter = ', SNR2, ' dB')
```

## Programming Exercise:

Check the statement $\left|Y(f)\right|^2 = \left|X(f)\right|^2 + \left|D(f)\right|^2$ by evaluating the relative error between the true value $\left|Y(f)\right|^2$ and the estimation $\left|X(f)\right|^2 + \left|D(f)\right|^2$. Use all columns and all rows of the power spectral density $\left|Y(f)\right|^2$ and $\left|X(f)\right|^2 + \left|D(f)\right|^2$ for evaluating the relative error.

```python
P_Y = None
idx1 = 0
idx2 = ws
while idx2 < y.shape[0]:
    # evaluate the spectra
    X = np.fft.fft(x[idx1:idx2] * w)
    D = np.fft.fft(d[idx1:idx2] * w)
    Y = np.fft.fft(y[idx1:idx2] * w)

    P_Y_local = (np.abs(Y)**2).reshape((X.shape[0], 1))
    P_Y_hat_local = (np.abs(X)**2 + np.abs(D)**2).reshape((X.shape[0], 1))
    if P_Y is None:
        P_Y = P_Y_local
        P_Y_hat = P_Y_hat_local
    else:
        P_Y = np.concatenate((P_Y, P_Y_local), axis = 1)
        P_Y_hat = np.concatenate((P_Y_hat, P_Y_hat_local), axis = 1)
    
    idx1 += hs
    idx2 += hs

RelativeSquaredError = np.sum((P_Y - P_Y_hat)**2) / np.sum(P_Y**2)
print('relative error = ', RelativeSquaredError)
```
