# Channel Estimation
A simple channel model assumes linear filtering without additive noise:

$y(n)=h(n)*x(n) = \sum_{m=0}^{M-1} h(m)\cdot x(n-m)$

In this jupyter notebook, an algorithm is introduced to reverse the effect of this simple channel:

$\tilde x(n)\approx y(n)*h^{-1}(n)$.

## Definition of input and output signal and filter
It is possible to estimate the impulse respone $h(n)$ by measuring the input signal $x(n)$ and the output signal $y(n)$.

In the following code, a simple impulse response $h(n)$ is assumed:

```python
import numpy as np
from scipy.linalg import solve_toeplitz, toeplitz
import matplotlib.pyplot as plt

DefaultImpulseResponseLength = 100

Fs = 48000
x = np.random.randn(Fs)
## Chanle parameters
a = 0.9
DelayInSamples = 10
h = np.zeros((DefaultImpulseResponseLength))
h[DelayInSamples] = 1.0
for n in range(DelayInSamples + 1, h.shape[0]):
    h[n] = a * h[n-1]
y = np.convolve(x, h)
y = y[:x.shape[0]]
```

## Estimation of Impulse Repsonse - Least Squares Approach

A simple approach is to estimate the impulse response by minimizing the squared error between the input signal $x(n)$ convolved with the estimated impulse response $\tilde h(n)$ and the output signal $y(n)$:

Minimize $\sum_n \left(\tilde h(n)*x(n) - y(n)\right)^2 = \sum_n \left(\left(\sum_{m=0}^{M-1}\tilde h(m)\cdot x(n-m)\right) - y(n)\right)^2$

Minimization is done, by setting the first derivative to zero:

$\frac{d}{dh(k)}=2\cdot\sum_n \left(\left(\sum_{m=0}^{M-1}\tilde h(m)\cdot x(n-m)\right) - y(n)\right)\cdot x(n-k)=0$

$\rightarrow\sum_n\sum_{m=0}^{M-1} \tilde h(m)\cdot x(n-m)\cdot x(n-k)=\sum_n y(n)\cdot x(n-k)$

$\rightarrow \sum_{m=0}^{M-1} \tilde h(m) \cdot \sum_n x(n-m)\cdot x(n-k) = \sum_n y(n)\cdot x(n-k)$

$\rightarrow \sum_{m=0}^{M-1} \tilde h(m)\cdot \varphi_{xx}(m-k) = \varphi_{xy}(k)$

```python
def CrossCorrelation(x, y = None):
    if y is None:
        FFTLength = int(2**np.ceil(1 + np.log2(x.shape[0])))
        Phi_xy = np.abs(np.fft.fft(x, FFTLength))**2
    else:
        MaximumLength = np.maximum(x.shape[0], y.shape[0])
        FFTLength = int(2**np.ceil(1 + np.log2(MaximumLength)))
        Phi_xy = np.conj(np.fft.fft(x, FFTLength)) * np.fft.fft(y, FFTLength)           
    return np.fft.ifft(Phi_xy).real

class CImpulseResponseEstimator(object):
    
    def __init__(self, LengthOfImpulseResponseInSamples = DefaultImpulseResponseLength):
        self.__LengthOfImpulseResponseInSamples = LengthOfImpulseResponseInSamples

    
    def Estimate(self, x, y):
        M = self.__LengthOfImpulseResponseInSamples
        assert M < x.shape[0], 'signal x must be longer than the impulse response'
        assert M < y.shape[0], 'signal x must be longer than the impulse response'
        phi_xx = CrossCorrelation(x)
        phi_xy = CrossCorrelation(x, y)
        h_est = solve_toeplitz(phi_xx[0:M], phi_xy[0:M])
        return h_est

AImpulseResponseEstimator = CImpulseResponseEstimator()
h_est = AImpulseResponseEstimator.Estimate(x, y)

y_est = np.convolve(x, h_est)
y_est = y_est[:x.shape[0]]
dB = 10*np.log10(np.sum(y**2) / (np.sum((y - y_est)**2) + 1e-10))
print('impulse response similarity: dB = ', dB)

plt.plot(h, 'b')
plt.plot(h_est, 'r')
plt.show()
```

## Estimation of Impulse Response - Normalized Least Mean Square (NLMS) Algorithm

For the NLMS algorithm, the difference between the local output sample $y(n)$ and the estimated local output sample $\tilde y(n)=\tilde h(n)*x(n)$ is derived:

$\frac{d}{dh(m)}\left(\tilde y(n) - y(n)\right)^2$

$=\frac{d}{dh(m)}\left(\left(\sum_{m=0}^{M-1} h(m)\cdot x(n-m)\right) - y(n)\right)^2$

$=2\cdot\left(\left(\sum_{m=0}^{M-1} h(m)\cdot x(n-m)\right) - y(n)\right)\cdot x(n-m)$

$=2\cdot\left(\tilde y(n) - y(n)\right)\cdot x(n-m)$

The estimation of the impulse response is updated by

$\tilde h(m) \leftarrow \tilde h(m) - \mu\cdot\left(\tilde y(n) - y(n)\right)\cdot x(n-m)$

with the stepsize $\mu$. So far, this algorithm is called Least Mean Square Algorithm.

It can be shown, that a normalization of the stepsize $\mu$ leads to good convergence behaviour. Therefore $\mu$ is replaced by:

$\frac{\mu}{\sum_{m=0}^{M-1} x^2(n-m)}$

```python
### implementation by ChatGPT
def nlms(x, y):
    mu = 1e-2
    N = len(x)
    h_est = np.zeros(DefaultImpulseResponseLength)  # Filter coefficients

    for n in range(DefaultImpulseResponseLength, N-1):
        x_n = x[n-DefaultImpulseResponseLength+1:n+1]  # Input vector
        y_hat = np.dot(h_est, x_n)  # Estimated output
        e = y[n] - y_hat  # Error signal
        norm_factor = np.dot(x_n, x_n) + 1e-10  # Small value to avoid division by zero
        h_est += (mu / norm_factor) * e * x_n  # Update filter coefficients

    return np.flip(h_est)

h_est = nlms(x, y)

y_est = np.convolve(x, h_est)
y_est = y_est[:x.shape[0]]

dB = 10*np.log10(np.sum(y**2) / (np.sum((y - y_est)**2) + 1e-10))
print('impulse response similarity: dB = ', dB)

plt.plot(h, 'b')
plt.plot(h_est, 'r')
plt.show()
```

## Inverse Filtering
In a typical measurement setup, $y(n)$ is the sensor signal, filtered by an impulse response $h(n)$. The original signal $x(n)$ is usually not accessable directly:

$y(n)=x(n)*h(n)$

An example could be: $x(n)$ is an acoustic noise produced at a certain position A of a machine. The sensor must be positioned at a different position B of the machine. The filtering between position A and B can be approximated by the FIR filter $h(n)$.

In order to evaluate the original signal $x(n)$ at position A, it may be possible to measure the output signal $y(n)$ for a certain input signal $x(n)$ under lab conditions.
In a second step, the proposed impulse response estimation algorithms can be used with $x(n)$ and $y(n)$ exchanged with each other. By this approach, an inverse impulse response $h^{-1}(n)$ is estimated:

$x(n)\approx y(n)*h^{-1}(n)$

With the knowledge of this inverse filtering under lab conditions, it should be possible, to reduce the impact of the filter $h(n)$ under real measurement scenarios.

One disadvantage of this approach is: delays cannot be inverted. Therefore, a separate delay estimation must be applied, and the impulse response must be estimated without this delay.

```python
def EvaluateCrossCorrelation(x, y):
    fftlen = np.maximum(x.shape[0], y.shape[0]) * 2
    phi = np.fft.ifft(np.fft.fft(y, fftlen) * np.conj(np.fft.fft(x, fftlen))).real
    return phi

def EstimateDelay(x, y):
    phi = EvaluateCrossCorrelation(x, y)
    T = np.argmax(np.abs(phi[0:(phi.shape[0]//2)]))
    return T

Delay_Estimated = EstimateDelay(x, y)
print('estimated delay in samples = ', Delay_Estimated)
h_inverse_est = AImpulseResponseEstimator.Estimate(y[Delay_Estimated:], x[:-Delay_Estimated])

plt.plot(h_inverse_est, 'k')
plt.title('estimated inverse impulse response')
plt.xlabel('samples')
plt.show()

x_est = np.convolve(y[Delay_Estimated:], h_inverse_est)
x_est = x_est[:x.shape[0]]

SNR = 10*np.log10(np.sum(x**2)/np.sum((x-x_est)**2))
print('SNR for inverse filtering = ', SNR, ', dB')
```

## Programming Exercise
Both proposed algorithms for estimating the impulse response ignore noise so far.
Evaluate the SNR between $x(n)$ and $\tilde y(n)$ for an additive noise of the channel in the range between $-10$ dB and $60$ dB.
Plot the resulting SNR of the least squares approach in blue and the normalised least mean squares approach in red over the SNR of the channel.

At which SNR is which algorithm performing better?

```python
TargetSNRList = np.arange(15) * 5 - 10
ResultingSNR = np.zeros((2, len(TargetSNRList)))
for TargetSNRIndex in range(len(TargetSNRList)):
    TargetSNR = TargetSNRList[TargetSNRIndex]
    d = np.random.randn(y.shape[0])
    noise_gain = np.sqrt(np.sum(y**2)/(np.sum(d**2)*(10**(TargetSNR/10))))
    z = y + noise_gain*d
    SNR = 10*np.log10(np.sum(y**2)/np.sum((z-y)**2))
    assert np.abs(SNR-TargetSNR) < 1e-2, 'wrong SNR'

    h_est = AImpulseResponseEstimator.Estimate(x, z)
    y_est = np.convolve(x, h_est)
    y_est = y_est[:x.shape[0]]
    ResultingSNR[0, TargetSNRIndex] = 10*np.log10(np.sum(y**2) / (np.sum((y - y_est)**2) + 1e-10))    

    h_est = nlms(x, y)
    y_est = np.convolve(x, h_est)
    y_est = y_est[:x.shape[0]]
    ResultingSNR[1, TargetSNRIndex] = 10*np.log10(np.sum(y**2) / (np.sum((y - y_est)**2) + 1e-10))        

plt.plot(TargetSNRList, ResultingSNR[0, :], 'b') # least squares
plt.plot(TargetSNRList, ResultingSNR[1, :], 'r') # normalized least mean squares
plt.show()
```
