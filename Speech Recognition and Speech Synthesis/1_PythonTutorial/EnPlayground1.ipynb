```python
import numpy as np

class CCircle(object):
    def __init__(self):
        self.__radius = 0

    def EvalArea(self):
        return np.pi*(self.__radius**2)

    def SetRadius(self, NewRadius):
        assert NewRadius >= 0.0, 'NewRadius >= 0'
        self.__radius = NewRadius
```

```python
r = 48000
f= 440
A = 1
w = 4
Delta = 2*A/(2**w)
t = np.arange(r)/r
x = A*np.sin(2*np.pi*f*t)
y = np.round(x/Delta)*Delta

import matplotlib.pyplot as plt

N = 100
plt.plot(t[:N], x[:N], 'b')
plt.plot(t[:N], y[:N], 'r')
plt.show()
```
