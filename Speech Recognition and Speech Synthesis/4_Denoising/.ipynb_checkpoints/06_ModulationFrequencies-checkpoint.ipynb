# Modulation Frequencies

```python
import matplotlib.pyplot as plt

import wave
import numpy as np
import os
os.chdir('../Python')

Filename = 'OldenburgerSatztest.wav'
with wave.open(Filename, 'rb') as wave_datei:
    rawdata = wave_datei.readframes(wave_datei.getnframes())
    Fs = wave_datei.getframerate()
    x = np.frombuffer(rawdata, dtype=np.int16) / (2**15)
```

```python
hs_in_ms = 10
hs = int(hs_in_ms * Fs / 1000)
ws = 2*hs
FFTLen = int(2**np.ceil(np.log2(ws)))
w = (1-np.cos(2*np.pi*(np.arange(ws)+0.5)/ws)) * 0.5
w = np.sqrt(w)
NyquistIndex = FFTLen // 2 + 1

def GetIndices(col):
    idx1 = col*hs
    idx2 = idx1+ws
    return idx1, idx2

def STFT(x):
    NumberOfColumns = (x.shape[0] - ws) // hs + 1
    X = np.zeros((NyquistIndex, NumberOfColumns), dtype=complex)
    for col in range(NumberOfColumns):
        idx1, idx2 = GetIndices(col)
        x_w = x[idx1:idx2] * w
        X[:, col] = np.fft.rfft(x_w, n=FFTLen)
    return X

def ISTFT(X):
    x = np.zeros((X.shape[1] * hs + ws - hs))
    for col in range(X.shape[1]):
        idx1, idx2 = GetIndices(col)
        x_w = np.fft.irfft(X[:, col])
        x[idx1:idx2] += x_w[0:ws] * w
    return x

x_tmp = np.copy(x)
X = STFT(x_tmp)
y_tmp = ISTFT(X)
x_tmp = x_tmp[hs:]
y_tmp = y_tmp[hs:-hs]
x_tmp = x_tmp[:y_tmp.shape[0]]
SNR = 10*np.log10(np.sum(x_tmp**2) / np.sum((x_tmp-y_tmp)**2))
assert SNR > 200, 'error in STFT / ISTFT'
```

```python
TargetSNR = 5
d = np.random.randn(x.shape[0])
noise_gain = np.sqrt(np.sum(x**2)/(np.sum(d**2)*(10**(TargetSNR/10))))
d *= noise_gain
y = x + d
RealSNR = 10*np.log10(np.sum(x**2) / np.sum((x-y)**2))
assert np.abs(RealSNR - TargetSNR) < 0.1, 'wrong noise level'
```

```python
X = STFT(x)
X_abs = np.abs(X)
X_ModulationFrequencies = np.fft.rfft(X_abs, axis = 1)
X_MF_abs = np.abs(X_ModulationFrequencies)

Y = STFT(y)
Y_abs = np.abs(Y)
Y_ModulationFrequencies = np.fft.rfft(Y_abs, axis = 1)
Y_MF_abs = np.abs(Y_ModulationFrequencies)

f = Fs * np.arange(Y.shape[0]) / FFTLen
ColumnSamplingFrequency = (hs_in_ms / 1000)**(-1)
f_MF = ColumnSamplingFrequency * np.arange(Y_MF_abs.shape[1]) / Y_abs.shape[1]
```

```python
SNR = 10*np.log10(X_MF_abs**2 / (np.abs(Y_MF_abs - X_MF_abs)**2))
SNR = np.maximum(SNR, 0.0)

plt.imshow(X_MF_abs, cmap='viridis', interpolation='nearest', extent=[f_MF[0], f_MF[-1],f[-1], f[0]], aspect=0.003)
plt.colorbar()
plt.show()

plt.imshow(SNR, cmap='viridis', interpolation='nearest', extent=[f_MF[0], f_MF[-1],f[-1], f[0]], aspect=0.003)
plt.colorbar()
plt.show()

plt.imshow(SNR, cmap='viridis', interpolation='nearest', extent=[f_MF[0], f_MF[-1],f[-1], f[0]], aspect=0.003)
plt.xlim(0.25, 25)
plt.ylim(8000, 125)

plt.colorbar()
plt.show()
```
